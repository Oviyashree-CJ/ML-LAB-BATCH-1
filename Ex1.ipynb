```python
import math
print("Floor value of 3.1428: ", math.floor(3.1428))
print("Ceil value of 3.1428: ", math.ceil(3.1428))
print("Square root of 25(in float): ", math.sqrt(25))
print("Square root of 25(in int): ", math.isqrt(25))
print("GCD of 3,5: ", math.gcd(3,5))
```

```python
import numpy

arr1 = numpy.array([1,5,3,2,4])
print("Array: ", arr1)
print("Array dimension: ", arr1.ndim)
print("Array shape: ", arr1.shape)
print("Array size: ", arr1.size)
print("Sum of array: ", arr1.sum())
print("Mean value: ",arr1.mean())
arr1.sort()
print("Sorted array: ", arr1)
print("Sine value: ", numpy.sin(arr1))
```

```python
from numpy.linalg import det, eig

arr2 = numpy.array([[5, 1], [3, 2]])
print("Determinant:", det(arr2))
e_val, e_vec = eig(arr2)
print("Eigenvalues:", e_val)
print("Eigenvectors:\n", e_vec)
```

```python
from scipy.linalg import det, eig

d_arr = numpy.array([[5, 1, 6], [3, 2, 0], [1, 3, 2]])
print("Array :\n",d_arr)
print("Determinent: ", det(d_arr))
```

```python
from scipy.linalg import det, eig

d_arr = numpy.array([[5, 1, 6], [3, 2, 0], [1, 3, 2]])
print("Array :\n",d_arr)
eigen_values, eigen_vectors = eig(d_arr)
print("Eigen values: ", eigen_values)
print("Eigen vectors: \n", eigen_vectors)
```

```python
import pandas as pd
info = {
    'Name': ['Arun', 'Kala', 'Jagan', 'Tina'],
    'Age': [18, 20, 17, 22],
    'Year': [2, 4, 1, 4],
    'Course': ['B.Arch', 'B.Tech', 'B.A', 'B.E']
}
df = pd.DataFrame(info)
melted_df = pd.melt(df, id_vars=['Name'], var_name='Attribute', value_name='Value')
print("Melted DataFrame:")
print(melted_df)
pivoted_df = melted_df.pivot(index='Name', columns='Attribute', values='Value').reset_index()
print("\nPivoted DataFrame:")
print(pivoted_df)
```
